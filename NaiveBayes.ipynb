```python
# Kondisi B dimana
# Day = Holiday
# Discount = Yes
# Free Delivery = Yes
# P(A) = NoBuy = PNo
# P(B) = PYes Discount, PYes Delivery, PHoliday
# P(B/A)1 = (PHoliday No|Buy) = PHolidayNo
# P(B/A)2 = (PYesDiscount|No Buy) = PYesDiscountNo
# P(B/A)3 = (PYesDelivery|No Buy) = PYesDeliveryNo
# P(B)1 = PHoliday
# P(B)2 = PYesDiscount
# P(B)3 = PYesDelivery

# ditanyakan P(A/B) ?
# Rumus
# P(A/B) = P(B|A)*P(A)/P(B)
```

```python
PNoBuy1 = ((PHolidayNo*PYesDiscountNo*PYesDeliveryNo)*PNo)/(PHoliday*PYesDiscount*PYesDelivery)
print(PNoBuy1)
```

```python
# Kondisi B dimana
# Day = Holiday
# Discount = Yes
# Free Delivery = Yes
# P(A) = Buy = PYes
# P(B) = PYes Discount, PYes Delivery, PHoliday
# P(B/A)1 = (PHoliday No|Buy) = PHolidayYes
# P(B/A)2 = (PYesDiscount|Buy) = PYesDiscountYes
# P(B/A)3 = (PYesDelivery|Buy) = PYesDeliveryYes
# P(B)1 = PHoliday
# P(B)2 = PYesDiscount
# P(B)3 = PYesDelivery

# ditanyakan P(A/B) ?
# Rumus
# P(A/B) = P(B|A)*P(A)/P(B)
```

```python
PBuy2 = ((PHolidayYes*PYesDiscountYes*PYesDeliveryYes)*PYes)/(PHoliday*PYesDiscount*PYesDelivery)
print(PBuy2)
```
